```python
def get_my_table():
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    columns = soup.select('th')
    columnlist=[]
    for column in columns:
        columnlist.append(column.text)
    df=pd.DataFrame(columns=columnlist)
    
    contents = soup.select('tr.row-data')
    
    dfcontent=[]
    alldfcontents=[]

    for content in contents:
        tds=content.find_all("td")
        for td in tds:
            dfcontent.append(td.text)
        alldfcontents.append(dfcontent)
        dfcontent=[]
        
    df = pd.DataFrame(columns = columnlist, data = alldfcontents)
    return df
```

```python
columnlist = ['No', '제목', '추천', '작성자', '조회', '작성일']
```

```python
total_df = pd.DataFrame(columns=columnlist)
total_df
```

```python
i = 0
while True:
    i += 1
    driver.get("http://www.ulsan.ac.kr/ubook/good_bookstory/recommend/index.aspx?pageNo="+ str(i)+ "&o=L")
    result = get_my_table()
    total_df = total_df.append(result)
```

```python
total_df = total_df.drop_duplicates(keep = "first")
```

```python
def get_my_table2():
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    title = soup.select("span.article-title")
    writer_company = soup.select("span.article-book-title")
    category = soup.select("span.category")
    who = soup.select("span.writer")
    count = soup.select("span.article-view-count")
    when = soup.select("span.article-register-date")
    
    
    columnlist = ['제목', '지은이_출판사', '분류', '작성자', '조회', '작성일']
    df=pd.DataFrame(columns=columnlist)
    
    temp = []
    for i in title:
        temp.append(i.text)
    df.제목 = temp
    
    temp = []
    temp2 = []
    for i in writer_company:
        temp.append(i.text)
    df.지은이_출판사 = temp
    
    temp = []
    for i in category:
        temp.append(i.text)
    df.분류= temp
    
    temp = []
    for i in who:
        temp.append(i.text.replace("\t", ''))
    df.작성자 = temp
    
    temp = []
    for i in count:
        temp.append(i.text)
    df.조회 = temp
    
    temp = []
    for i in when:
        temp.append(i.text)
    df.작성일 = temp
    return df
```
