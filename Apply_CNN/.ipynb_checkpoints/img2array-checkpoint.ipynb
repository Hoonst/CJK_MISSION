```python
def img2array(country, numImg):
    '''
    input : fname 1~300.jpg
    return : np.array of fnames
    '''
    for i in range(numImg):
        pil_im = pil.open('./'+ country +'/' + str(i)+ '.jpg')
        print('The size of the image is ', pil_im.size)
```
