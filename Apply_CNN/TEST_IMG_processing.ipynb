# 1번째 사진

```python
pil_im = pil.open('./psy.jpg')
print('The size of the image is ', pil_im.size)
```

```python
pil_im
```

```python
w, h = pil_im.size
w, h
pil_im = pil_im.crop((0, 170, w, h-190)) #left, top, right,bottom
pil_im
```

```python
pil_im = pil_im.resize((80,80), pil.ANTIALIAS)
pil_im
```

```python
pil_im.save('psy_cropped1.jpg')
```

```python
fnames = ['psy_cropped1.jpg','psy_cropped2.jpg', 'pkr_cropped.jpg', 'nyj_cropped.jpg']
```

```python
friends = np.random.randint(low=1,high=2,size=(1,80,80,3))

for fname in fnames:
    pil_im = pil.open('./'+ fname)
    friends = np.vstack((friends, np.array(pil_im)[np.newaxis,]))
    
friends = friends[1:,]
friends = np.transpose(np.array(friends), axes = (0,3,1,2))
print('The shape of \'friends\' array is ' , friends.shape)
friends
```

```python
np.savetxt('friends.txt', friends.flatten(), fmt='%d')
friends2 = np.loadtxt('friends.txt', dtype=int)
assert np.all(friends == friends2.reshape(4,3,80,80))
```
